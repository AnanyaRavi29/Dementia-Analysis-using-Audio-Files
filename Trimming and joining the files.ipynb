```python

from pydub import AudioSegment
import pandas as pd

def separate_and_combine_audio(input_audio_file, timestamps_file, output_file1, output_file2):

    audio = AudioSegment.from_file(input_audio_file)


    timestamps_df = pd.read_csv(timestamps_file)


    speaker1_audio = AudioSegment.empty()
    speaker2_audio = AudioSegment.empty()

    Total_time = 0


    for index, row in timestamps_df.iterrows():
        start_time = row['start'] * 1000
        end_time = row['end'] * 1000
        print(start_time, end_time)
        Total_time += end_time - start_time

        speaker_audio = audio[start_time:end_time]


        if row['speaker'] == 'SPEAKER_00':
            speaker1_audio += speaker_audio
        elif row['speaker'] == 'SPEAKER_01':
            speaker2_audio += speaker_audio


    speaker1_audio.export(output_file1, format="wav")
    speaker2_audio.export(output_file2, format="wav")

    print("Total time speakers were speaking:",Total_time/1000,"s")

audio_file = "/content/drive/MyDrive/Dementia Dataset/ADReSSo21-diagnosis-train/ADReSSo21/diagnosis/train/audio/ad/adrso024.wav"
timestamps_file = "/content/drive/My Drive/Dementia Dataset/Diarized csv files/diarization_segments.csv"
output_file1 = "/content/drive/MyDrive/Dementia Dataset/Trimmed audio files_via code/speaker1_output.wav"
output_file2 = "/content/drive/MyDrive/Dementia Dataset/Trimmed audio files_via code/speaker2_output.wav"
separate_and_combine_audio(audio_file, timestamps_file, output_file1, output_file2)



```

```python
import os

# Directory containing audio files
input_directory = "/content/drive/MyDrive/Dementia Dataset/ADReSSo21-diagnosis-train/ADReSSo21/diagnosis/train/audio/"

# Directory to save output files
output_directory = "/content/drive/MyDrive/Dementia Dataset/Trimmed audio files_via code/"

# Directory containing timestamps files
timestamps_directory = "/content/drive/My Drive/Dementia Dataset/Diarized csv files/"

# List all audio files in the input directory
audio_files = os.listdir(input_directory)

# Iterate over each audio file
for audio_file in audio_files:
    # Check if the file is a WAV file
    if audio_file.endswith(".wav"):
        # Construct the full path of the audio file
        audio_path = os.path.join(input_directory, audio_file)

        # Construct the full path of the timestamps file
        timestamps_file = os.path.join(timestamps_directory, audio_file.replace(".wav", "_diarization_segments.csv"))

        # Construct the output file paths
        output_file1 = os.path.join(output_directory, "speaker1_" + audio_file)
        output_file2 = os.path.join(output_directory, "speaker2_" + audio_file)

        # Apply the separate_and_combine_audio function
        separate_and_combine_audio(audio_path, timestamps_file, output_file1, output_file2)
```
